```python
def count_ingredients(row):
    ing = list(row)[1:]
    count = len([x for x in ing if x != ''])
    return count

count_ingredients(equipment.iloc[4])
```

```python
for x in equipment.columns[1:]:
    equipment[x] = equipment[x].apply(fold)

equipment["ingredient_count"] = equipment.apply(count_ingredients, axis=1)

equipment.sort_values(by=['ingredient_count'], ascending=False, inplace=True)

equipment.sample(10)
```

```python
allcols = []
for x in equipment.columns[1:-1]:
    allcols.extend(equipment[x].unique())

messy_uniques = list(set(allcols))
messy_uniques.sort(key=str.lower)
print(len(messy_uniques))
messy_uniques
```

If we want to do a reverse lookup so that we can see which recipes use a piece of equipment, we can do that here:

```python
def row_contains(row, values_to_look_for):
    values_in_row = list(row[1:])
    for v in values_to_look_for:
        it_does = v in values_in_row
        if it_does:
            return it_does
    return False

on_hand = ['Fine-mesh sieve']
uses = equipment.apply(lambda x: row_contains(x, on_hand), axis=1)
these_recipes = equipment[uses]
list(these_recipes["title"])
```

Which bit of kit gets used in the most recipes?

```python
counts = {}

for thing in messy_uniques:
    on_hand = [thing]
    uses = equipment.apply(lambda x: row_contains(x, on_hand), axis=1)
    these_recipes = equipment[uses]
    counts[thing] = {"count": these_recipes.shape[0], "recipes": list(these_recipes["title"])}
    

c = pd.DataFrame.from_dict(counts, orient="index")
c.sort_values(by=['count'], ascending=False, inplace=True)
c[c["count"]>1]
```
